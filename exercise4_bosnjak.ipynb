```python
#linear interpolation is a method to determine an unknown point x between two known points (x0, y0) and (x1, y1)
#this we do by assuming a linear function from (x0, y0) to (x1, y1) on which x lays
#linear interpolation: y(x) = y0 + [(y1 - y0)/(x1 - x0)] * (x - x0)

#the step size is the distance between two known points, between which you built a linear interpolation 
#step size: d = x_i+1 - x_i

#the maximum error is related to the second derivative of the function 
#E = (y"(x) * d^2)/8 

#first i wanna calculate the second derivative of the function 
def y(x):
    return math.pi * (x ** 7)

def first_derivative(x):
    return math.pi * 7 * (x ** 6)

def second_derivative(x):
    return math.pi * 7 * 6 * (x ** 5)

#now i have to set the limit for my absolute error 
E_max_lim = 10 ** (-6) 

#to find the step size d i use the formula d = sqrt(8E/max|y"(x)|)
#i first have to define a random intervall for x
random_intervall = np.arange(0, 1, 0.1)
#now i calculate the maximum value of my second derivative in the random intervall 
#second_derivative_values = np.array([second_derivative for x in random_intervall])
max_value = np.max(second_derivative(random_intervall)) 


d = np.sqrt((8 * E_max_lim) / max_value)

print(d)  
```
